## Exploratory Data Analysis and Visualization

```r
# Calculate proportion of missing values in each column
na_prop <- wildfire %>%
    summarise(across(everything(), ~round(mean(is.na(.)), 4)))
t(na_prop)
```

Two variables, `ia_arrival_at_fire_date` and `fire_fighting_start_date`, are missing more than 20% of their values. Because of this, they will be removed from the analysis.

```r
wildfire_clean <- wildfire %>% select(-ia_arrival_at_fire_date, -fire_fighting_start_date)
```

```r
# Select all the categorical variables to check for class imbalance
wildfire_class_check <- wildfire %>% select(where(is.character), -fire_number)
head(wildfire_class_check)
```

```r
# Check for class imbalance with pie charts of each categorical variable
for (i in 1:ncol(wildfire_class_check)){
    prop <- prop.table(table(wildfire_class_check[,i]))
    title <- names(wildfire_class_check)[[i]]
    pie(prop, main = title)
    }
```

There are a few categorical variables that seem to exhibit class imbalance, including `size_class`, `responsible_group`, `fire_type`, `fire_position_on_slope`, and `ia_access`.
